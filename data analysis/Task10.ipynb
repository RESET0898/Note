```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('CDNOW.xlsx',names=['用户ID', '购买日期','购买数量', '购买金额'],index=['key'])
data.head()
```

```python
data.groupby('用户ID').购买金额.mean()#各用户ID的购买金额的平均数
```

```python
data.groupby('用户ID').购买金额.describe()#每个用户ID的购买金额的描述型统计
```

```python
data.groupby('用户ID').购买金额.agg(['max','min'])
```

```python
import matplotlib.pyplot as plt
```

```python
data['购买费用'] = data['购买数量'] * data['购买金额']
```

```python
data.head()
```

```python
total = data.groupby('用户ID').购买费用.sum()
total
```

```python
total.plot.bar()
```

```python
plt.xlabel('用户ID')
plt.ylabel('购买总额')
plt.title('各用户ID的购买总额')
plt.show()
```
